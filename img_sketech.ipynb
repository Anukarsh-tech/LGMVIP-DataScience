here we have imported the cv2 library

```python
import cv2
from google.colab.patches import cv2_imshow
```

Reading the imgage in "img varibale

```python
img=cv2.imread("marvel.jpg")
cv2_imshow(img)
```

Converting the real image to gray color image 


```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

Transforming the gray image to inverted image

```python
inverted_img=255-gray_img
cv2_imshow(inverted_img)
```

Now at final the sketched image

```python
blurred=cv2.GaussianBlur(inverted_img,(21,21),0)
```

```python
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray_img,inverted_blurred,scale=256.0)
cv2_imshow(pencil_sketch)
```

Show Real image and Sketced image

```python
cv2_imshow(img)
cv2_imshow(pencil_sketch)
```
