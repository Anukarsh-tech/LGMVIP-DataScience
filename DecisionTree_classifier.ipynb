# shuffling of data

```python
iris=iris.sample(frac=1)
```

# Encoding of Target data

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
label_encode=LabelEncoder()
```

```python
iris["Encode_Sepcies"]=label_encode.fit_transform(iris["Species"])
```

```python
iris.head(20)
```

# converting label column into an array

```python
en=label_encode.fit_transform(iris["Species"])
```

```python
en 
```

# Importing the tain_test_split 

```python
from sklearn.model_selection import train_test_split
```

```python
x=iris.iloc[:,:-2].values
```

```python
print(x)
```

```python
print(x.shape)
```

```python
y=en
```

```python
y
```

# Spliting the dataset into train and test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
                                             
```

```python
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

# importing the DecisionTreeClassifier from sklearn.Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

  
  # Fitting the training data

```python
dtc.fit(x_train,y_train)
```

# predicting the test datset

```python
predict=dtc.predict(x_test)
```

```python
predict
```

# importing the accuracy_score and checking the accuracy of model

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,predict)
```

```python
acc
```

# Visulaizing the predicted results through confusion_matrix and heatmap

```python
from sklearn.metrics import confusion_matrix
```

```python
conf=confusion_matrix(y_test,predict)
conf
```

```python
sns.heatmap(conf,annot=True)
```
